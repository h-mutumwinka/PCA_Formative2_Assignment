### Step 1: Load and Standardize the Data
Before applying PCA, we must standardize the dataset. Standardization ensures that all features have a mean of 0 and a standard deviation of 1, which is essential for PCA.
Fill in the code to standardize the dataset.

STRICTLY - Write code that implements standardization based on the image below




```python
# Step 1: Load and Standardize the data (use of numpy only allowed)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('https://raw.githubusercontent.com/h-mutumwinka/PCA_Formative2_Assignment/refs/heads/main/DatasetAfricaMalaria.csv')

print("Initial dataset shape:", data.shape)

data = data.select_dtypes(include=['float64', 'int64'])
data = data.dropna()
data.head()

X = data.values
mean = np.mean(X, axis=0)
std = np.std(X, axis=0)
standardized_data = (X - mean) / std
standardized_data[:5]
```

### Step 3: Calculate the Covariance Matrix
The covariance matrix helps us understand how the features are related to each other. It is a key component in PCA.

```python
# Step 3: Calculate the Covariance Matrix
cov_matrix = np.cov(standardized_data, rowvar=False)
cov_matrix
```

### Step 4: Perform Eigendecomposition
Eigendecomposition of the covariance matrix will give us the eigenvalues and eigenvectors, which are essential for PCA.
Fill in the code to compute the eigenvalues and eigenvectors of the covariance matrix.

```python
# Step 4: Perform Eigendecomposition
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
eigenvalues, eigenvectors
```

### Step 5: Sort Principal Components
Sort the eigenvectors based on their corresponding eigenvalues in descending order. The higher the eigenvalue, the more important the eigenvector.
Complete the code to sort the eigenvectors and print the sorted components.

<a url ='https://www.youtube.com/watch?v=vaF-1xUEXsA&t=17s'>How Is Explained Variance Used In PCA?'<a/>

```python
# Step 5: Sort Principal Components

sorted_indices = np.argsort(eigenvalues)[::-1]

sorted_eigenvectors = eigenvectors[:, sorted_indices]

sorted_eigenvalues = eigenvalues[sorted_indices]


explained_variance_ratio = sorted_eigenvalues / np.sum(sorted_eigenvalues)

cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

print("Explained variance ratio by principal components:")
for i, var in enumerate(explained_variance_ratio[:5]):
    print(f"PC{i+1}: {var:.4f} ({cumulative_variance_ratio[i]:.4f} cumulative)")

sorted_eigenvectors
```

### Step 6: Project Data onto Principal Components
Now that weve selected the number of components, we will project the original data onto the chosen principal components.
Fill in the code to perform the projection.

```python
# Step 6: Project Data onto Principal Components

num_components = 2 # Decide on the number of principal components to keep
#I picked the first two principal components because they catch most of the action in the data.
#Its like keeping the biggest, juiciest pieces of a cake and
#throwing away the tiny crumbsstill yummy and nothing important is lost
reduced_data = np.dot(standardized_data, sorted_eigenvectors[:, :num_components])
reduced_data[:5]
```

### Step 7: Output the Reduced Data
Finally, display the reduced data obtained by projecting the original dataset onto the selected principal components.

```python
# Step 7: Output the Reduced Data
print(f'Reducing the Data Shape: {reduced_data.shape}')  # Display reduced data shape
reduced_data[:5]  # Display the first 5 rows of reduced data
```

### Step 8: Visualize Before and After PCA
Now, let's plot the original data and the data after PCA to compare the reduction in dimensions visually.

```python
# Step 8: Visualize Before and After PCA
plt.figure(figsize=(14, 6))

# Plot original data (first two features for simplicity)
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], alpha=0.7)
plt.title('Original Data (First Two Features)')
plt.xlabel('Original Data Feature 1')
plt.ylabel('Original Data Feature 2')
plt.grid(True)

# Plot reduced data after PCA
plt.subplot(1, 2, 2)
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], alpha=0.7, c='r')
plt.title('Data After PCA ')
plt.xlabel('Original Principal Component 1')
plt.ylabel('Original Principal Component 2')
plt.grid(True)

plt.tight_layout()
plt.show()

#After PCA, we look at the data in a new way. The dots stay the same,
#but the lines change direction. PC1 shows the biggest change in the data,
#so it keeps the most important information.
```
